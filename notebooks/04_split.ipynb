```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_parquet("./data//train_features.parquet")
```

```python
print(df['dt'].min(), df['dt'].max())
```

```python
# TIme based split
# Train : first 70% dates
# val : next : 15% dates
# test : final : 15% dates

dates = df['dt'].sort_values().unique()
n = len(dates)

train_end = dates[int(n*0.70)]
val_end = dates[int(n*0.85)]

print(f"Total unique dates: {n}")
print(f"Train end:  {train_end}")
print(f"Val end:    {val_end}")
```

```python
# Split data based on data boundaries
train_df = df[df['dt']<=train_end]
val_df = df[(df['dt'] > train_end) & (df['dt']<=val_end)]
test_df = df[df['dt']> val_end]
```

```python
print(f"Train: {train_df.shape} | {train_df['dt'].min().date()} to {train_df['dt'].max().date()}")
print(f"Val:   {val_df.shape}   | {val_df['dt'].min().date()} to {val_df['dt'].max().date()}")
print(f"Test:  {test_df.shape}  | {test_df['dt'].min().date()} to {test_df['dt'].max().date()}")
```

```python
train_df.to_parquet("../data/processed/train_split.parquet", index=False)
val_df.to_parquet("../data/processed/val_split.parquet", index=False)
test_df.to_parquet("../data/processed/test_split.parquet", index=False)
```
