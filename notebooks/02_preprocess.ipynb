```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_parquet("./data/train.parquet")
```

```python
# Parsing date and sorting it

df['dt'] = pd.to_datetime(df['dt'])
df = df.sort_values(['store_id','product_id','dt']).reset_index(drop=True)
```

```python
print(df.dtypes[['dt', 'store_id', 'product_id']])
print(df.head(3)[['store_id', 'product_id', 'dt', 'sale_amount']])
```

```python
# Creating a stockout flag feature
df['stockout_flag'] = (df['stock_hour6_22_cnt'] >= 2).astype(int)
```

```python
print(df['stockout_flag'].value_counts())
print(f"\nStockout rate: {df['stockout_flag'].mean()*100:.1f}%")
```

```python
keep_cols = [
    'store_id', 'product_id', 'city_id',
    'first_category_id', 'second_category_id', 'third_category_id',
    'dt', 'sale_amount', 'stock_hour6_22_cnt', 'stockout_flag',
    'discount', 'holiday_flag', 'activity_flag',
    'precpt', 'avg_temperature', 'avg_humidity', 'avg_wind_level'
]
```

```python
df = df[keep_cols]
print(df.shape)
print(df.columns.tolist())
```

```python
df.to_parquet(".data/interim/train_clean.parquet", index=False)
```
